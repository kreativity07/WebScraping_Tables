# Special Codes for Olympics

Use Pandas to scrape the following site and decode the special codes used for Olympics since 1896 until 2018.

```python
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_IOC_country_codes'
oly_codes = ['Code', 'Nation/Team', 'Years', 'Notes']
```

```python
# Use Panda's `read_html` to parse the url
tables = pd.read_html(url)
tables
```

```python
df = tables[4]
df
```

```python
df.columns = ['Code', 'Nation/Team', 'Years', 'Notes']
df
```

```python
# Drop the first row and set the index to the `Code` column
df = df.iloc[1:]
df.set_index('Code', inplace=True)
df
```

```python
# Loop through the list of olympic codes and print the abbreviation
# along with the full description from the `Nation/Team` column.
# Use the DataFrame to perform the lookup.
for Code in df.index.values:
    print(Code, df.loc[Code]['Nation/Team'])
```
